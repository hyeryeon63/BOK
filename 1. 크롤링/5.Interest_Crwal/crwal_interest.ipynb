```python
import requests
from bs4 import BeautifulSoup
kosdaq_result=[]
for i in range(1,697):
    
    url=f"https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page={i}"
    resp=requests.get(url)
    soup=BeautifulSoup(resp.text)
    li=soup.select('div.section_exchange2 tr')
    
    for l in li:
        if l.find('td',class_='date') is None:
            continue
        kosdaq_result.append({
            'date':l.find_all('td')[0].text.strip(),
            'num':l.find_all('td')[1].text,
            'updown':l.find_all('td')[3].text.strip()
        })
kosdaq_result
```

```python
import pandas as pd
p=pd.DataFrame(kosdaq_result)
```
